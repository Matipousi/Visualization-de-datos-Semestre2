```python
df2 = pd.read_csv("Ventas2017 Suc_nº2.csv", encoding="latin1")
df2.head()
df2.columns
```

```python
df3 = pd.read_csv("Ventas2018 Suc_nº1.csv", encoding="latin1")
df3.head()
df3.columns
```

Vamos a agregar el nombre de la sucursal en la base de 2017

```python
df2['Sucursal'] = 'Sucursal_n2'
df2.head()
```

Unimos en una sola base los 3 df anteriores

```python
# Unimos verticalmente
df_total = pd.concat([df1, df2, df3], ignore_index=True)
# con ignore_index=True, no conserva los índices originales de los DataFrames sino que crea un nuevo índice secuencial desde 0 hasta el final del DataFrame combinado
df_total
```

```python
df_total.info() #explorando
```

Como tenemos varios vaores nulos en la columna Monto Venta, vamos a calcular este valor usando la operacion de multiplicación (unidades * precio)

```python
df_total['Monto Venta'] = df_total['Unidades Vendidas'] * df_total['Precio Venta']
df_total
```

Ahora ajustemos la fecha, en los valores que tenemos texto en mes

```python
#Código para extraer el valor entre las barras /
df_total['Mes_Texto'] = df_total['Fecha'].str.extract(r'/([^/]+)/')
df_total
```

```python
df_total['Mes_Texto'].value_counts()
#df_total['Mes_Texto'].value_counts().sort_index()
```

```python
# Diccionario para reemplazar los meses en texto por números
meses_dict = {'Set.-': '09','Oc.': '10','Nov.-': '11','Di.': '12'}

# Reemplazar en la misma columna Fecha
for mes_texto, mes_num in meses_dict.items():
    df_total['Fecha'] = df_total['Fecha'].str.replace(mes_texto, mes_num, regex=False)
```

```python
df_total
```

```python
df_total
```

No necesitamos la última columna, por lo que la vamos a borrar.

```python
#Elimino la columna Mes_Texto ya que no me interesa mantenerla
df_total.drop(columns=['Mes_Texto'], inplace=True)
#inplace=True modifica el DataFrame original sin necesidad de reasignarlo.
df_total
```

```python
#df_total.head()           # Ver primeras filas
#df_total.columns          # Ver nombres de variables
#df_total.info()           # Estructura del DataFrame
df_total[['Unidades Vendidas', 'Precio Venta']].describe()     # Estadísticas generales
```
